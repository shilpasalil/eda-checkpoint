## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
#creating visualization using scatter plot with rank of the colleges vs median base salary

ax=sns.scatterplot(x="rank", y="medianBaseSalary",data=df)
ax.set_title("rank and medianBaseSalary")
plt.show()
```

```python
#creating visualization using heat map for correlation between the rank of the colleges vs total student population


plt.title("correlation between the rank and totalStudentPop ")
sns.heatmap(df[["rank","totalStudentPop"]].corr())
plt.show()
```

```python

import matplotlib.pyplot as plt
from matplotlib import style
# setting style for graphs
style.use('ggplot')
plt.rcParams['figure.figsize'] = (20,10)
import matplotlib.colors as mcolors

#line plot

plt.plot(df["rank"],df["percentOfStudentsFinAid"], color ='#0c06f7')
plt.title('analyze percentOfStudentsFinAid by rank', color = 'black', fontsize = 16, fontweight = 'bold')
plt.xlabel('rank',color = 'black',fontsize = 16, fontweight = 'bold')
plt.ylabel('percentOfStudentsFinAid',color = 'black',fontsize = 16, fontweight = 'bold')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.show()

#--------------------------
#scatter graph

ax=sns.scatterplot(x="rank", y="percentOfStudentsFinAid",data=df)
ax.set_title("analyze rank and percentOfStudentsFinAid")
plt.show()
```

```python
#analyze public/private schools in different states
#bar graph
plt.bar(df["collegeType"], df["state"], color='#894585')
#plt.bar(keys, values, color='light eggplant') - need code for color
plt.xlabel('College typr', color = 'black', fontsize = 16, fontweight = 'bold')
plt.ylabel('State', color = 'black', fontsize = 16, fontweight = 'bold')
plt.title('types of collegen in different states', color = 'black', fontsize = 16, fontweight = 'bold')
plt.show()


# #-----
# #scatter plot
# ax=sns.scatterplot(x="rank", y="percentOfStudentsFinAid",data=df)
# ax.set_title("analyze rank and percentOfStudentsFinAid")
# plt.show()
```

```python
#college count per region

plt.figure(figsize=(12,7))
sns.histplot(data=df, x='region') 
plt.show()
```

```python
#creating a vizualiztion of US map using latitude and longitude,
#where we can hower over different states to show lat, long, student pop, and rank
#color coded by rank giving an easyunderstanding
#size of bubbles on studentPopulation
 
#reference to learn about Plotly Express(px)
#https://plotly.com/python/px-arguments/#:~:text=Plotly%20Express%20is%20the%20easy,bar%20or%20px.
#documentationfor scatter_geo()
#https://plotly.com/python-api-reference/generated/plotly.express.scatter_geo
#documentation for update_layout
#
    
    
import plotly.express as px
viz1 = px.scatter_geo(df, lat='latitude', lon='longitude',
                     color='rank', hover_name='state',size='studentPopulation',
                     )
viz1.update_layout(
    title='<b>Colleges in the USA colored by rank and size of plot by student population</b>',
    geo_scope='usa'
#     geo=dict(
#         scope='usa')
)
viz1.show()

```

```python

#compare public and private schools in different states
import plotly.express as px
viz1 = px.scatter_geo(df, lat='latitude', lon='longitude',
                     color='collegeType', hover_name='state',
                     )
viz1.update_layout(
    title='<b>public or private colleges in different states</b>',
    geo_scope='usa'
#     geo=dict(
#         scope='usa')
)
viz1.show()


#*********************************************
#We will divide the data into two tables, Private and Public, and compare them
# df['collegeType']=df['collegeType'].str.replace('-',' ')
# Private=df.collegeType=='Private not for profit'
# Public=df.collegeType=='Public'
# print("private:\n-----------------------\n",df[Private].count())
# print("-----------------------\npublic:\n-----------------------\n",df[Public].count())
```
